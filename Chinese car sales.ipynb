# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
Y= df['price']
X= scaler.fit_transform(df.drop(["price","car_ID"],axis=1))
X= pd.DataFrame(data=X,columns = df.drop(columns=['price','car_ID']).columns)
X.head()
```

```python
X.corr()
```

```python
## pair of independent ariables with correlation greater than 0.5 
k=X.corr()
## syntax
##Result=[(element of final ist) x (how to take different values) x (condition to store the final element)]
z=[[str(i),str(j)] for i in k.columns for j in k.columns if(k.loc[i,j]>abs(0.5))&(i!=j)]
z,len(z)
```

# Calculating variance inflation factor

```python
#importing varience inflation factor from the stats model
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif_data=X

## calculating VIF for every column

VIF=pd.Series([variance_inflation_factor(vif_data.values,i) for i in range(vif_data.shape[1])],index=vif_data.columns)
VIF
```

# Treating Multicollinearity

```python
def MC_remover(data):
    vif=pd.Series([variance_inflation_factor(data.values,i)for i in range(data.shape[1])],index=data.columns)
    if vif.max()>5:
        print(vif[vif == vif.max()].index[0],'has been removed')
        data = data.drop(columns=[vif[vif==vif.max()].index[0]])
        return data
    else:
        print("No multicollinearity present anymore")
        return data
```

```python
for i in range(10):
    vif_data=MC_remover(vif_data)
vif_data.head()
    
```

```python
## calculating VIF for remaining columns
VIF=pd.Series([variance_inflation_factor(vif_data.values,i) for i in range(vif_data.shape[1])],index=vif_data.columns)
VIF,len(vif_data.columns)
```

# Feature Selection

```python
x=vif_data
y=df['price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=101)

x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

# Prediction Using Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression(normalize =True)
lr.fit(x_train,y_train)
```

```python
lr.coef_ #output is all the coeffients of the independent variables such as mo,m1,m2,m3,m4....
#as data is normalized so value of intercept mo=0
```

```python
prediction=lr.predict(x_test)
prediction
```

```python
lr.score(x_test,y_test)
```

# Model coefficients

```python
coefficients_table=pd.DataFrame({'column':x_train.columns,
                                'coefficients':lr.coef_})
coefficients_table=coefficients_table.sort_values(by="coefficients")
```

```python
plt.figure(figsize=(8,6),dpi=120)
x=coefficients_table['column']
y=coefficients_table['coefficients']
plt.barh(x,y)
plt.xlabel("coefficients")
plt.ylabel("variables")
plt.title("Normalized coefficients")
```

#### observations:

1. The variables in the above graph which have positive values are directly proportional to price and which has negative values are indirectly proportional to car price
